```python
import pandas as pd
import altair as alt

# Leer el archivo CSV con (;)
df = pd.read_csv('Radio2021Listo.csv', sep=';', encoding='latin1')

# Limpiar nombres de columnas
df.columns = df.columns.str.strip().str.lower()

# Para contar géneros
genero_counts = df['genero'].value_counts().reset_index()
genero_counts.columns = ['genero', 'cantidad']

# Crear gráfico
chart = alt.Chart(genero_counts).mark_bar(color='#1f77b4').encode(
    x=alt.X('cantidad:Q', title='Cantidad de canciones'),
    y=alt.Y('genero:N', sort='-x', title='Género'),
    tooltip=['genero:N', 'cantidad:Q']
).properties(
    width=600,
    height=400,
    title='Cantidad de canciones por género en Radio 2021'
).configure_axis(
    labelFontSize=12,
    titleFontSize=14
).configure_title(
    fontSize=18,
    anchor='start'
).configure_view(
    stroke=None
)

# Guardar como HTML
chart.save('genero_radio_2021.html')

# Mostrar el gráfico
chart
```

```python
import pandas as pd
import altair as alt

# Leer Archivos y años
archivos = {
    2021: 'Radio2021Listo.csv',
    2022: 'Radio2022Listo.csv',
    2023: 'Radio2023Listo.csv',
    2024: 'Radio2024Listo.csv',
}

# Leer y preparar datos para artistas
dfs_artistas = []
for año, archivo in archivos.items():
    df = pd.read_csv(archivo, sep=';', encoding='latin1')
    df['Artista'] = df['Artista'].str.strip().str.title()
    df['Año'] = año
    dfs_artistas.append(df[['Artista', 'Año']])

# Juntar todos los años
df_artistas = pd.concat(dfs_artistas)

# Contar canciones por artista y año
artista_counts = df_artistas.groupby(['Año', 'Artista']).size().reset_index(name='cantidad')

# Definir top N
top_n = 10

# Obtener top N artistas por año
top_artistas_por_año = (
    artista_counts.sort_values(['Año', 'cantidad'], ascending=[True, False])
    .groupby('Año')
    .head(top_n)
    .reset_index(drop=True)
)

# Crear gráfico comparativo de artistas por año
chart_artistas = alt.Chart(top_artistas_por_año).mark_bar().encode(
    x=alt.X('cantidad:Q', title='Cantidad de canciones'),
    y=alt.Y('Artista:N', sort='-x', title='Artista'),
    color=alt.Color('Año:N', title='Año', scale=alt.Scale(scheme='category10')),
    column=alt.Column('Año:N', title=None),
    tooltip=['Artista:N', 'cantidad:Q', 'Año:N']
).properties(
    width=200,
    height=300,
    title='Top 10 artistas con más canciones por año (2021-2024)'
).configure_axis(
    labelFontSize=10,
    titleFontSize=12
).configure_title(
    fontSize=16,
    anchor='start'
)

chart_artistas

chart.save('grafico_radio_2021-2024_chile.html')


# Mostrar el gráfico
chart

```

```python
import pandas as pd
import altair as alt

archivos = {
    2021: 'Radio2021Listo.csv',
    2022: 'Radio2022Listo.csv',
    2023: 'Radio2023Listo.csv',
    2024: 'Radio2024Listo.csv',
}

datos = []

for año, archivo in archivos.items():
    df = pd.read_csv(archivo, sep=';', encoding='latin1')
    df.rename(columns=lambda x: x.strip(), inplace=True)

    df['Nacionalidad'] = df['Nacionalidad'].str.strip().str.upper()

    conteos = df['Nacionalidad'].value_counts().reset_index()
    conteos.columns = ['Nacionalidad', 'Cantidad']
    conteos['Año'] = año

    datos.append(conteos)

df_nacionalidades = pd.concat(datos, ignore_index=True)

# Calcular top 10 nacionalidades con más canciones sumando todos los años
top10_nacionalidades = (
    df_nacionalidades.groupby('Nacionalidad')['Cantidad']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .index.tolist()
)

# Filtrar solo las nacionalidades top 10 para el gráfico
df_top10 = df_nacionalidades[df_nacionalidades['Nacionalidad'].isin(top10_nacionalidades)]

# Crear gráfico
chart = alt.Chart(df_top10).mark_bar().encode(
    x=alt.X('Año:O', title='Año'),
    y=alt.Y('Cantidad:Q', title='Cantidad de canciones'),
    color=alt.Color('Nacionalidad:N', title='Nacionalidad'),
    tooltip=['Nacionalidad', 'Cantidad', 'Año']
).properties(
    width=700,
    height=400,
    title='Top 10 Nacionalidades en ranking de Radio (2021-2024)'
).configure_axis(
    labelFontSize=12,
    titleFontSize=14
).configure_title(
    fontSize=18,
    anchor='start'
)

chart
chart.save('graficonacion_radio_2021_chile.html')
```

```python
import pandas as pd
import altair as alt

# Cargar archivo de Spotify 2021
spotify = pd.read_csv('/content/Spotify2021Listo (1) nacho.csv', sep=';', encoding='latin1')

# Limpiar nombres de columnas
spotify.columns = spotify.columns.str.strip()
spotify.rename(columns={'Genero ': 'Genero'}, inplace=True)

# Normalizar texto de géneros a minúsculas para facilitar la comparación
spotify['Genero'] = spotify['Genero'].str.lower()

# Filtrar canciones urbanas o reggaetón (puedes ajustar los términos clave)
generos_urbanos = ['urbano', 'reggaetón', 'reggaeton', 'trap']  # Añade más si es necesario
es_urbano = spotify['Genero'].apply(lambda g: any(term in g for term in generos_urbanos))

# Calcular cantidades
total_canciones = len(spotify)
canciones_urbanas = es_urbano.sum()
porcentaje = canciones_urbanas / total_canciones * 100

# Crear DataFrame para gráfico
df_resultado = pd.DataFrame({
    'Tipo': ['Urbano/Reggaetón', 'Otros'],
    'Cantidad': [canciones_urbanas, total_canciones - canciones_urbanas]
})

# Crear gráfico de pastel o barras
chart = alt.Chart(df_resultado).mark_arc(innerRadius=50).encode(
    theta=alt.Theta(field='Cantidad', type='quantitative'),
    color=alt.Color(field='Tipo', type='nominal'),
    tooltip=['Tipo', 'Cantidad']
).properties(
    title=f'Participación de música urbana/reggaetón en Spotify Chile 2021 ({porcentaje:.1f}%)'
)

chart
chart.save('porcentaje_urbano_spotify_2021.html')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
df = pd.read_csv('/content/Spotify2021Listo (1) nacho.csv', sep=';', on_bad_lines='skip')
df.columns = df.columns.str.strip()

# Filtrar nacionalidades distintas de Chile (excluir Chile aunque esté combinado)
df = df[~df['Nacionalidad'].str.lower().str.contains('chile')]

# Dividir combinaciones múltiples de nacionalidades (por coma)
# y expandir filas (explode) para contar individualmente
df['Nacionalidad'] = df['Nacionalidad'].str.upper()  # uniformar
df['Nacionalidad'] = df['Nacionalidad'].str.replace(' ', '')  # quitar espacios
df_exploded = df.assign(Nacionalidad=df['Nacionalidad'].str.split(',')).explode('Nacionalidad')

# Contar ocurrencias por nacionalidad
conteo = df_exploded['Nacionalidad'].value_counts().reset_index()
conteo.columns = ['Nacionalidad', 'Cantidad']

# Guardar HTML
conteo.to_html('/content/estesi_spotify_chile_limpio.html')
print("Archivo HTML generado correctamente.")

# Gráfico
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.barplot(data=conteo, x='Cantidad', y='Nacionalidad', palette='viridis')
plt.title('Nacionalidades dominantes en ranking Spotify Chile (excluyendo Chile)', fontsize=14)
plt.xlabel('Cantidad de posiciones en el ranking')
plt.ylabel('Nacionalidad')
plt.tight_layout()
plt.show()
```

```python
# Instalar unidecode si no está instalado
!pip install unidecode

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import unidecode
from google.colab import files

# Función para limpiar nombres de artistas
def limpiar_artista(nombre):
    nombre = str(nombre)
    nombre = nombre.upper()
    nombre = unidecode.unidecode(nombre)
    nombre = nombre.replace("FT.", "").replace("FEAT.", "").replace("FEAT", "").replace("FT", "")
    nombre = nombre.replace("&", "Y")
    nombre = nombre.strip()
    return nombre

# Cargar archivos
df_spotify = pd.read_csv('/content/Spotify2022Listo (1) nacho.csv', sep=';', on_bad_lines='skip')
df_radios = pd.read_csv('/content/Radio2022Listo.csv', sep=';', on_bad_lines='skip')

# Limpiar columnas
df_spotify.columns = df_spotify.columns.str.strip()
df_radios.columns = df_radios.columns.str.strip()

# Normalizar textos
df_spotify['Artista'] = df_spotify['Artista'].apply(limpiar_artista)
df_spotify['Nacionalidad'] = df_spotify['Nacionalidad'].str.upper().str.strip()

df_radios['Artista'] = df_radios['Artista'].apply(limpiar_artista)
df_radios['Nacionalidad'] = df_radios['Nacionalidad'].str.upper().str.strip()

# Filtrar artistas chilenos
chilenos_spotify = df_spotify[df_spotify['Nacionalidad'].str.contains('CHILE')]
chilenos_radios = df_radios[df_radios['Nacionalidad'].str.contains('CHILE')]

# Crear sets
artistas_spotify = set(chilenos_spotify['Artista'])
artistas_radios = set(chilenos_radios['Artista'])

# Comparaciones
en_ambos = sorted(artistas_spotify & artistas_radios)
solo_spotify = sorted(artistas_spotify - artistas_radios)
solo_radios = sorted(artistas_radios - artistas_spotify)

# Mostrar resumen
print(f"✅ Artistas en ambos: {len(en_ambos)}")
print(f"🎵 Solo en Spotify: {len(solo_spotify)}")
print(f"📻 Solo en Radios: {len(solo_radios)}")

# Guardar resultados como HTML
pd.DataFrame(en_ambos, columns=["Artista"]).to_html('/content/artistas_chilenos_en_ambos_2022.html', index=False)
pd.DataFrame(solo_spotify, columns=["Artista"]).to_html('/content/artistas_chilenos_solo_spotify_2022.html', index=False)
pd.DataFrame(solo_radios, columns=["Artista"]).to_html('/content/artistas_chilenos_solo_radios_2022.html', index=False)

# Descargar resultados
files.download('/content/artistas_chilenos_en_ambos_2022.html')
files.download('/content/artistas_chilenos_solo_spotify_2022.html')
files.download('/content/artistas_chilenos_solo_radios_2022.html')
```

```python
import pandas as pd
import altair as alt
import os
import re

# Función para extraer el año desde la ruta del archivo
def agregar_año_desde_ruta(df, ruta):
    try:
        filename = os.path.basename(ruta)
        match = re.search(r'\d{4}', filename)
        if match:
            año = int(match.group(0))
            df['Año'] = año
    except Exception as e:
        print(f"Error al extraer año de {ruta}: {e}")
    return df

# Función para cargar y limpiar cada archivo
def cargar_y_limpiar(ruta, medio):
    try:
        df = pd.read_csv(ruta, encoding='latin1', sep=';', engine='python', on_bad_lines='skip')
        df.columns = df.columns.str.strip().str.replace('ï»¿', '', regex=False).str.replace('AÃ±o', 'Año', regex=False)

        renombres = {
            'Genero ': 'Genero',
            'Año ': 'Año',
            'Posición ': 'Posición',
            'Artista ': 'Artista',
            'Nacionalidad ': 'Nacionalidad'
        }
        df = df.rename(columns={k: v for k, v in renombres.items() if k in df.columns})

        df['Medio'] = medio
        df = agregar_año_desde_ruta(df, ruta)

        for col in ['Genero', 'Artista', 'Nacionalidad']:
            if col in df.columns:
                df[col] = df[col].astype(str).str.strip()

        if 'Posición' in df.columns:
            df['Posición'] = pd.to_numeric(df['Posición'], errors='coerce')
            df = df.dropna(subset=['Posición'])

        return df
    except Exception as e:
        print(f"Error procesando archivo {ruta}: {e}")
        return None

# Cargar todos los archivos
archivos = {
    'Radio': ['/content/Radio2021Listo.csv', '/content/Radio2022Listo.csv'],
    'Spotify': ['/content/Spotify2021Listo (1) nacho.csv', '/content/Spotify2022Listo (1) nacho.csv']
}

dfs = []
for medio, rutas in archivos.items():
    for ruta in rutas:
        df_proc = cargar_y_limpiar(ruta, medio)
        if df_proc is not None:
            dfs.append(df_proc)

# Unir todo en un solo DataFrame
if dfs:
    df = pd.concat(dfs, ignore_index=True)
    df['Año'] = pd.to_numeric(df['Año'], errors='coerce')
    df = df[df['Año'].isin([2021, 2022])]
    df = df.dropna(subset=['Año'])
    df['Genero'] = df['Genero'].astype(str).str.strip()

    print("Columnas del DataFrame:", df.columns.tolist())
    display(df.head())

    # Agrupar por Año y Medio para contar géneros distintos
    df_genres = df.groupby(['Año', 'Medio'])['Genero'].nunique().reset_index()
    df_genres = df_genres.rename(columns={'Genero': 'Generos_distintos'})

    # Crear gráfico con Altair
    chart = alt.Chart(df_genres).mark_bar().encode(
        x=alt.X('Medio:N', title='Medio'),
        y=alt.Y('Generos_distintos:Q', title='Cantidad de géneros distintos'),
        color='Medio:N',
        column=alt.Column('Año:N', title='Año')
    ).properties(
        title='Diversidad de géneros por medio (Radio vs Spotify)',
        width=150
    )

    # Guardar gráfico como archivo HTML
    chart.save('diversidad_generos_medio.html')

else:
    print("No se pudieron cargar los DataFrames.")
```

```python
import pandas as pd
import altair as alt
import os
import re

# Función para extraer año desde la ruta
def agregar_año_desde_ruta(df, ruta):
    try:
        filename = os.path.basename(ruta)
        match = re.search(r'\d{4}', filename)
        if match:
            año = int(match.group(0))
            df['Año'] = int(match.group(0))
    except:
        pass
    return df

# Función para cargar y limpiar cada archivo
def cargar_y_limpiar(ruta, medio):
    try:
        df = pd.read_csv(ruta, encoding='latin1', sep=';', engine='python', on_bad_lines='skip')
        df.columns = df.columns.str.strip().str.replace('ï»¿', '', regex=False).str.replace('AÃ±o', 'Año', regex=False)
        df = df.rename(columns=lambda x: x.strip())

        renombres = {
            'Genero ': 'Genero',
            'Año ': 'Año',
            'Posición ': 'Posición',
            'Artista ': 'Artista',
            'Nacionalidad ': 'Nacionalidad'
        }
        df = df.rename(columns={k: v for k, v in renombres.items() if k in df.columns})
        df['Medio'] = medio
        df = agregar_año_desde_ruta(df, ruta)

        for col in ['Genero', 'Artista', 'Nacionalidad']:
            if col in df.columns:
                df[col] = df[col].astype(str).str.strip()

        if 'Posición' in df.columns:
            df['Posición'] = pd.to_numeric(df['Posición'], errors='coerce')
            df = df.dropna(subset=['Posición'])

        return df
    except Exception as e:
        print(f"Error procesando archivo {ruta}: {e}")
        return None

# Archivos por medio
archivos = {
    'Radio': ['/content/Radio2021Listo.csv', '/content/Radio2022Listo.csv'],
    'Spotify': ['/content/Spotify2021Listo (1) nacho.csv', '/content/Spotify2022Listo (1) nacho.csv']
}

# Cargar todos
dfs = []
for medio, rutas in archivos.items():
    for ruta in rutas:
        df_proc = cargar_y_limpiar(ruta, medio)
        if df_proc is not None:
            dfs.append(df_proc)

# Unir todos los DataFrames
df = pd.concat(dfs, ignore_index=True)

# Limpiar artistas
df['Artista'] = df['Artista'].str.strip().str.lower()

# Crear sets
artistas_spotify = set(df[df['Medio'] == 'Spotify']['Artista'].unique())
artistas_radio = set(df[df['Medio'] == 'Radio']['Artista'].unique())

# Clasificar artistas
artistas_data = []

for artista in artistas_spotify.union(artistas_radio):
    if artista in artistas_spotify and artista in artistas_radio:
        origen = 'Ambos'
    elif artista in artistas_spotify:
        origen = 'Solo Spotify'
    else:
        origen = 'Solo Radio'
    artistas_data.append({'Artista': artista.title(), 'Origen': origen})

df_artistas = pd.DataFrame(artistas_data)

# Crear gráfico interactivo
chart = alt.Chart(df_artistas).mark_bar().encode(
    x=alt.X('count():Q', title='Cantidad de artistas'),
    y=alt.Y('Origen:N', title='Presencia en medios', sort='-x'),
    color='Origen:N',
    tooltip=['Origen:N', 'count():Q']
).properties(
    title='Distribución de artistas por medio',
    width=400,
    height=200
)

# Crear tabla interactiva
tabla = alt.Chart(df_artistas).mark_text(align='left').encode(
    y=alt.Y('Artista:N', sort=alt.EncodingSortField('Artista', order='ascending')),
    text='Artista:N',
    tooltip=['Origen:N']
).transform_filter(
    alt.datum.Origen != 'Ambos'
).properties(
    width=300,
    height=8000,
    title='Artistas únicos por medio (excluye "Ambos")'
)

# Combinar gráfico + tabla
final = alt.vconcat(chart, tabla).configure_title(anchor='start')

# Guardar como HTML
final.save('artistas_repetidos_y_unicos.html')

print("✅ Archivo guardado como 'artistas_repetidos_y_unicos.html'")
```
