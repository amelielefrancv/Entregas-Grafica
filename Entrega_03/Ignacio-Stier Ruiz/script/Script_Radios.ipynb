```python
import pandas as pd
import altair as alt
df = pd.read_csv("Radio2021Listo.csv", sep=";", encoding="latin1", engine="python")

df.columns = df.columns.str.strip().str.lower()

df.head()
```

```python
import pandas as pd
import altair as alt


df = pd.read_csv("Radio2021Listo.csv", encoding='latin1', sep=';')


df.columns = df.columns.str.strip()


genre_by_year = df.groupby(['Año', 'Genero']).size().reset_index(name='Cuenta')


genre_by_year['Porcentaje'] = genre_by_year.groupby('Año')['Cuenta'].transform(lambda x: x / x.sum() * 100)


input_dropdown = alt.binding_select(options=sorted(genre_by_year['Año'].unique()))
selection = alt.selection_point(fields=['Año'], bind=input_dropdown, name='Selecciona el año')


pie = alt.Chart(genre_by_year).mark_arc().encode(
    theta=alt.Theta(field='Porcentaje', type='quantitative'),
    color=alt.Color(field='Genero', type='nominal'),
    tooltip=['Genero', 'Cuenta', alt.Tooltip('Porcentaje', format='.1f')]
).add_params(
    selection
).transform_filter(
    selection
).properties(
    width=400,
    height=400,
    title='Proporción de géneros musicales por año en Radio durante 2021'
)


pie.show()
pie.save("grafico_generos_radio_2021.html")
```
