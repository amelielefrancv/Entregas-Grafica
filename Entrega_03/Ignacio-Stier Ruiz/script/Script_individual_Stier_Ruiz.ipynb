```python
import pandas as pd
import altair as alt
df= pd.read_csv("Spotify2021Listo.csv")
```

```python
df
```

```python
alt.Chart(df) .mark_bar(bandSize=10).encode(
   x="Posición",
   y="Genero"
)
```

```python
genero_counts = (
    df.groupby("Genero")
      .size()
      .reset_index(name="Cantidad")
)

# Luego de tener ya subida la base de datos del archivo, lo dividimos de tal forma de mostar los porcentajes de las canciones escuchadas
total = genero_counts["Cantidad"].sum()
genero_counts["Porcentaje"] = 100 * genero_counts["Cantidad"] / total

# Despues hacemos el gráfico de torta
pie = (
    alt.Chart(genero_counts)
       .mark_arc()
       .encode(
           theta="Cantidad:Q",
           color="Genero:N",
           tooltip=[
               alt.Tooltip("Genero:N", title="Género"),
               alt.Tooltip("Cantidad:Q", title="Cantidad"),
               alt.Tooltip("Porcentaje:Q", format=".1f", title="Porcentaje (%)")
           ]
       )
       .properties(
           width=400,
           height=400,
           title="Distribución de géneros escuchados en Spotify – 2024"
       )
)

pie
```

```python
pie.save('grafico_generos_spotify_2024.html')
```
